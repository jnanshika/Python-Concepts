#generator to generate n even numbers

```python
def even_numbers(n):
    count = 0
    num = 0
    while count < n:
        yield num
        num += 2
        count += 1


for even in even_numbers(5):
    print(even)

```

```python
def generator():
    yield 1
    yield 2
    yield 3

gen = generator()
print(gen)
print("After for loop")
for i in generator():
    print(i)
```

```python
#using generator depict the functionality of range
def gen(n):
    i = 0.0
    while i < n:
        yield i 
        i += 1.0

for i in gen(10):
    print(i)

#more advance look
def frange(start, stop, step):
    current = start
    while (step > 0 and current < stop) or (step < 0 and current > stop):
        yield round(current, 10)  # rounding to handle floating-point precision
        current += step

for i in frange(0.0, 10.0, 1.0):
    print(i)
```

```python
#Create a generator that produces an infinite sequence of Fibonacci numbers.
def gen(num):
    yield 1
    old = 1
    count = 1
    while count <= num:
        yield old
        old += count
        count+=1

for i in gen(5):
    print(i)
```
